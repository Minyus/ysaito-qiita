## Simulation

```python
means = [0.10, 0.30, 0.50, 0.70, 0.90]
n_arms = len(means)
random.shuffle(means)
arms = pd.Series(map(lambda x: BernoulliArm(x), means))
```

### EpsilonGreedy

```python
eg1 = EpsilonGreedy(0.2, [], [])
eg2 = EpsilonGreedy(0.5, [], [])
```

```python
eg1.initialize(n_arms)
eg2.initialize(n_arms)
```

```python
result1 = test_algorithm(eg1, arms, 1000, 1000)
result2 = test_algorithm(eg2, arms, 1000, 1000)
```

```python
df1 = to_frame(result1)
df2 = to_frame(result2)
```

### SoftMax

```python
sm1 = SoftMax(0.2, [], [])
sm2 = SoftMax(0.5, [], [])
```

```python
sm1.initialize(n_arms)
sm2.initialize(n_arms)
```

```python
result3 = test_algorithm(sm1, arms, 1000, 1000)
result4 = test_algorithm(sm2, arms, 1000, 1000)
```

```python
df3 = to_frame(result3)
df4 = to_frame(result4)
```

### UCB1

```python
ucb1 = UCB1([], [])
```

```python
ucb1.initialize(n_arms)
```

```python
result5 = test_algorithm(ucb1, arms, 1000, 1000)
```

```python
df5 = to_frame(result5)
```

## 結果の描画

```python
average_rewards([df1, df2, df3, df4, df5],  
                ["Epsilon Greedy : 0.2", "Epsilon Greedy :0.5", 
                 "SoftMax : 0.2", "SoftMax : 0.5", "UCB"])
```
