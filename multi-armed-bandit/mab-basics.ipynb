## Simulation 1

### Arms

```python
means1 = [0.10, 0.30, 0.50, 0.70, 0.90]
random.shuffle(means1)
arms1 = pd.Series(map(lambda x: BernoulliArm(x), means1))
```

### Algorithmns

```python
algos1 = [EpsilonGreedy(0.2, len(arms1)), # ε - greedy (ε = 0.2)
          EpsilonGreedy(0.5, len(arms1)), # ε - greedy (ε = 0.5)
          SoftMax(0.2, len(arms1)), # Softmax(τ = 0.2)
          SoftMax(0.5, len(arms1)), # Softmax(τ = 0.5)
          UCB1(len(arms1))] # Upper Confidence Bound 1
```

### Run

```python
result_list = sim_mabs(algos1, arms1, 1000, 1000)
```

### Visualize

```python
average_rewards(result_list,  
                ["Epsilon Greedy : 0.2", "Epsilon Greedy :0.5", 
                 "SoftMax : 0.2", "SoftMax : 0.5", "UCB"])
```
